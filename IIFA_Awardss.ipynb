```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url="https://en.wikipedia.org/wiki/International_Indian_Film_Academy_Awards"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
awards_data=soup.find(class_="sortable wikitable")
awards_data
```

```python
all_rows=awards_data.find_all('tr')
len(all_rows)
```

```python
heads=[d.get_text().strip()for d in all_rows[0].find_all('th')]
heads
```

```python
#award=[d.find().get_text().strip() for d in all_rows[1:]]
# award=[]
# for d in all_rows[1:]:
#     L1=[]
#     for h in d.find_all('td'):
#         L1.append(h.get_text().strip())
#     award.append(L1)

awards = [[h.get_text().strip() for h in d.find_all('td')] for d in all_rows[1:]]
awards
```

```python
for r in awards:
    print(len(r))
```

```python
d={k:[] for k in heads}
for row in awards:
    if len(row)==8:
        d['Ceremony'].append(row[0])
        d['Date'].append(row[1])
        d['Best Film winner'].append(row[2])
        d['Host(s)'].append(row[3])
        d['Venue'].append(row[4])
        d['City'].append(row[5])
        d['Country'].append(row[6])
        d['Continent'].append(row[7])
        cont=row[7]
        
    elif len(row)==7:
        d['Ceremony'].append(row[0])
        d['Date'].append(row[1])
        d['Best Film winner'].append(row[2])
        d['Host(s)'].append(row[3])
        d['Venue'].append(row[4])
        d['City'].append(row[5])
        d['Country'].append(row[6])
        d['Continent'].append(cont)
        host=row[3]
        
    elif len(row)==6:
        
        d['Ceremony'].append(row[0])
        d['Date'].append(row[1])
        d['Best Film winner'].append(row[2])
        d['Host(s)'].append(host)
        d['Venue'].append(row[3])
        d['City'].append(row[4])
        d['Country'].append(row[5])
        d['Continent'].append(cont)
        
    else:
        d['Ceremony'].append(row[0])
        d['Date'].append(row[1])
        d['Best Film winner'].append(row[2])
        d['Host(s)'].append(row[3])
        d['Venue'].append(row[3])
        d['City'].append(row[3])
        d['Country'].append(row[3])
        d['Continent'].append(row[3])

d        
```

```python
df=pd.DataFrame(d)
df
```
